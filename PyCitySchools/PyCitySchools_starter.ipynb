```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

# initial checks
#school_data_complete['Student ID'].notnull().value_counts()
#school_data_complete.shape
#school_data_complete.dtypes
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = len(school_data_complete.groupby(['school_name']))

#Calculate the total number of students 
total_students = len(school_data_complete.groupby(['Student ID']))

#Calculate the total budget
total_budget = school_data_complete['budget'].sum()

#Calculate the average math score
avg_math_score = school_data_complete['math_score'].mean()

# Calculate the average reading score
avg_reading_score = school_data_complete['reading_score'].mean()

#Calculate the percentage of students with a passing math score (70 or greater)
students_math_above70 =len(school_data_complete.loc[school_data_complete['math_score']>=70,:])/ total_students *100

#Calculate the percentage of students with a passing reading score (70 or greater)
students_reading_above70 = len(school_data_complete.loc[school_data_complete['reading_score']>=70,:])/ total_students *100

#Calculate the percentage of students who passed math and reading (% Overall Passing)
percentage_overall_passing =len(school_data_complete.loc[(school_data_complete['reading_score']>=70) & 
                            (school_data_complete['math_score']>=70) ,:])/ total_students *100

#Create a dataframe to hold the above results
keys = pd.Series(['total schools','total students','total budget','average math score','average readingscore','students passing on math %','students passing on reading %','Overall Passing %'])
values =pd.Series([total_schools,total_students,total_budget,avg_math_score, avg_reading_score,students_math_above70, students_reading_above70,percentage_overall_passing])

district_summary = pd.DataFrame(columns=[keys,values]).astype(float)
district_summary.round(3)
district_summary.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School name
school_name = school_data_complete['school_name'].unique()
school_name = pd.Series(school_name)
#print(school_name)

#School type
school_type = school_data[['school_name','type']].set_index('school_name')
#print(school_type)

#Count Students per School
school_students = school_data_complete.groupby(['school_name'])['Student ID'].count()
school_students=school_students
#print(school_students)

#Total School Budget
school_budget = school_data_complete.groupby(['school_name'])['budget'].max()
school_budget=school_budget

#print(school_budget)

# Per student budget
student_budget = school_budget/school_students
student_budget=student_budget
#print(student_budget)

#Average Math Score
avg_math = school_data_complete.groupby(['school_name'])['math_score'].mean()
#print(avg_math)

#Average Reading Score
avg_reading = school_data_complete.groupby(['school_name'])['reading_score'].mean()
#print(avg_reading)

# % passing math
math_above70 = school_data_complete.loc[school_data_complete['math_score'] >= 70,:]
pass_math = math_above70.groupby(['school_name'])['math_score'].count() / school_students * 100
#print(pass_math)

# % passing reading
reading_above70 = school_data_complete.loc[school_data_complete['reading_score'] >= 70,:]
pass_reading = reading_above70.groupby(['school_name'])['reading_score'].count() / school_students * 100
#print(pass_reading)

#% Overall Passing
overall_above70 = school_data_complete.loc[(school_data_complete['reading_score']>=70) & 
                            (school_data_complete['math_score']>=70) ,["school_name","Student ID"]]
pass_overall = overall_above70.groupby(['school_name'])['Student ID'].count() / school_students * 100
#print(pass_overall)


# Create a data frame
school_summary = pd.concat([school_type, school_students, student_budget,avg_math,avg_reading,pass_math,
                            pass_reading,pass_overall], axis=1)


school_summary.columns.values[0] = "School Type"
school_summary.columns.values[1] = "Total Students"
school_summary.columns.values[2] = "Per Student Budget"
school_summary.columns.values[3] = "Average Math Score"
school_summary.columns.values[4] = "Average Reading Score"
school_summary.columns.values[5] = "% Passing Math"
school_summary.columns.values[6] = "% Passing Reading"
school_summary.columns.values[7] = "% Overall Passing"
school_summary.head(15)


```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#checks number of grades in the data frame
grades = len(school_data_complete['grade'].unique())
#print(grades)

#Create a pandas series for each grade
grade_9th = school_data_complete.loc[school_data_complete['grade']=='9th',:]
grade_10th = school_data_complete.loc[school_data_complete['grade']=='10th',:]
grade_11th = school_data_complete.loc[school_data_complete['grade']=='11th',:]
grade_12th = school_data_complete.loc[school_data_complete['grade']=='12th',:]
#print(grade_9th)


#Group each series by school and calculates average of math score
grade_9th_byschool = grade_9th.groupby(['school_name'])['math_score'].mean()
#print(grade_9th_byschool)
grade_10th_byschool = grade_10th.groupby(['school_name'])['math_score'].mean()
#print(grade_10th_byschool)
grade_11th_byschool = grade_11th.groupby(['school_name'])['math_score'].mean()
#print(grade_11th_byschool)
grade_12th_byschool = grade_12th.groupby(['school_name'])['math_score'].mean()
#print(grade_12th_byschool)

#Math Scores by Grade dataframe
math_bygrade = pd.merge(grade_9th_byschool, grade_10th_byschool, on="school_name")
math_bygrade = math_bygrade.rename(columns={'math_score_x':'9th','math_score_y':'10th'})
#print(math_bygrade)
math_bygrade = pd.merge(math_bygrade, grade_11th_byschool, on="school_name")
#print(math_bygrade)
math_bygrade = pd.merge(math_bygrade, grade_12th_byschool, on="school_name")
#print(math_bygrade)
math_bygrade = math_bygrade.rename(columns={'math_score_x':'11th','math_score_y':'12th'})
math_bygrade.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0, 584.9999,629.9999,644.9999,679.9999]
group_names = ['<$585','$585-630','$630-645','$645-680']

#bins data and append a column to school summary
school_summary['Spending Ranges (Per Student)']=pd.cut(school_summary["Per Student Budget"], bins, labels=group_names)

#creates new series that are average of the columns at school_summary grouped by the new column appended on the step above
school_spending_math = school_summary.groupby('Spending Ranges (Per Student)')['Average Math Score'].mean()
#print(school_spending_math)
school_spending_reading = school_summary.groupby('Spending Ranges (Per Student)')['Average Reading Score'].mean()
#print(school_spending_reading)
school_spending_passmath = school_summary.groupby('Spending Ranges (Per Student)')['% Passing Math'].mean()
#print(school_spending_passmath)
school_spending_passreading = school_summary.groupby('Spending Ranges (Per Student)')['% Passing Reading'].mean()
#print(school_spending_passreading)
school_spending_pass = school_summary.groupby('Spending Ranges (Per Student)')['% Overall Passing'].mean()
#print(school_spending_pass)

#concatenates new series in a new dataframe
school_spending = pd.concat([school_spending_math,school_spending_reading,school_spending_passmath,school_spending_passreading,school_spending_pass],axis=1)
school_spending = round(school_spending,2)
school_spending.head()
```
